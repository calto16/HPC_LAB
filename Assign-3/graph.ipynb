```python
import matplotlib.pyplot as plt

# Data
matrix_sizes = [500, 1000, 2000, 10000]
threads = [2, 4, 8, 16]

# Times (in seconds) for each matrix size and number of threads
times = {
    500: [0.000543, 0.000440, 0.000452, 0.000607],
    1000: [0.002010, 0.001512, 0.001022, 0.001414],
    2000: [0.006306, 0.003910, 0.003646, 0.004124],
    10000: [0.407307, 0.285126, 0.204092, 0.181238]
}

# Plot
plt.figure(figsize=(10, 6))

for size in matrix_sizes:
    plt.plot(threads, times[size], label=f'Matrix Size: {size}')

plt.xlabel('Number of Threads')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime vs. Number of Threads for Different Matrix Sizes')
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
